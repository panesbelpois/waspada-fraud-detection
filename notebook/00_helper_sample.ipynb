# Helper — Ambil Contoh Transaksi dari Dataset
Jalanin notebook ini kalau butuh data contoh buat testing API / frontend.

```python
import pandas as pd
import json

df = pd.read_csv('creditcard.csv')
```

## Contoh Transaksi FRAUD

```python
# Ambil 5 transaksi fraud
fraud_samples = df[df['Class'] == 1].head(5)

for i, (_, row) in enumerate(fraud_samples.iterrows()):
    data = {k: v for k, v in row.to_dict().items() if k != 'Class'}
    print(f'=== FRAUD #{i+1} ===')
    print(json.dumps(data, indent=2))
    print()
```

## Contoh Transaksi NORMAL

```python
# Ambil 5 transaksi normal
normal_samples = df[df['Class'] == 0].head(5)

for i, (_, row) in enumerate(normal_samples.iterrows()):
    data = {k: v for k, v in row.to_dict().items() if k != 'Class'}
    print(f'=== NORMAL #{i+1} ===')
    print(json.dumps(data, indent=2))
    print()
```
